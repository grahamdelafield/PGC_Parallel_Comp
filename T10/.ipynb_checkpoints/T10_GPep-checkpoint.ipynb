```python
# find the glycopeptides unique to C18 and PGC

def create_ident(dataframe, columns):
    '''
    This function creates a unique string that combines
    each characteristic provided in columns.
    
    :param(dataframe): (pd.DataFrame) dataframe containing Byonic output
    :param(columns): (list) columns to be combined

    Ex: 
        df = peptide seq  |  glycan  |  charge
             ---------------------------------
              PEPTIDER   HexNax(2)Hex(5)  2
        
    >>> create_ident(['peptide seq', 'glycan', 'charge'])
    >>> 'PEPTIDER_HexNax(2)Hex(5)_2'
    '''
    err_str =  'Only list-like items accepted in func create_ident()'
    assert isinstance(columns, list), err_str

    sub = dataframe[columns]
    for col in sub:
        sub.loc[:, col] = sub[col].astype(str)
    sub['res'] = sub[columns].apply(list, axis=1)
    return ['_'.join(item) for item in sub.res]

def num_sacch(s):
    '''
    This function counts the number of listed glucose
    units according to a byonic output:
    
    :param(s): (str) string of glycan
    
    Ex: s = 'HexNAc(2)Hex(5)'
        >>> num_sacch(s)
        >>> 7 
    '''
    pattern = re.compile(r'[0-9]+')
    match = re.findall(pattern, s)
    match = [int(i) for i in match]
    return sum(match)

all_gpeps = {}
for file in files:
    base = ntpath.basename(file)
    base, _ = ntpath.splitext(base)
    print(f'Parsing file {file}')
    bf = ByFile(file)
    bf.fill_no_glycans()
    bf.remove_reverse(modify=True)
    bf.determine_glycosites()
    bf.filter_hits(modify=True, score=100)
    bf.frame = bf.reduce_frame(gp_only=True)
    total = bf.total_gp()
    total.loc[:, 'source'] = [base]*len(total)
    all_gpeps[base] = set(create_ident(total, ['clean_peptide', 'glycan', 'z']))
    
keys = [
    [k for k in all_gpeps.keys() if re.search('C18', k)],
    [k for k in all_gpeps.keys() if re.search('PGC', k)],
]
```

```python
def combine_dfs(d1, d2):
    if d1.empty:
        return d2
    else:
        return pd.concat([d1, d2], axis=0) 
```

```python
# average length of backbone for unique glycopeptides
groups = [pgc, c18]
names = ['pgc', 'c18']
labels = []
char = []
values = []
for group in groups:
    other = groups[len(groups)-1 - groups.index(group)]
    other_name = names[len(groups)-1 - groups.index(group)]
    peps = []
    tot = []
    sac = []
    zs = []
    rat = []
    nc = []
    for p in group:
        if p not in other:
            tot.append(p)
            back, glyc, z = p.split('_')
            peps.append(len(back))
            sac.append(num_sacch(glyc))
            zs.append(int(z))
            nc.append(int(z) / len(back))
            rat.append(num_sacch(glyc) / len(back))
    values.append(len(tot)) # num unique
    values.append(np.mean(peps)) # avg length
    values.append(np.mean(sac)) # num glucose
    values.append(np.mean(zs)) # avg charge
    values.append(np.mean(nc))
    values.append(np.mean(rat)) # glycan to peptide length ratio
    char.extend(['Unique Glycopeptides', 'Avg. Length', 'Num G.U.', 'Avg. Charge', 'Norm Charge', 'G:P Ratio'])
    labels.extend([names[groups.index(group)]]*6)

shared_gp = pd.DataFrame({'column':labels, 'type':char, 'value':values})


from altair.expr import datum

base = alt.Chart(shared_gp).mark_bar().encode(
    x=alt.X('column:O', title=''),
    y=alt.Y('value:Q', title=''),
    column='type:O',
    color='column:O'
)
chart = alt.hconcat()
for t in shared_gp.type.unique():
    chart |= base.transform_filter(datum.type == t)
chart
```

```python
keys = [
    [k for k in all_gpeps.keys() if re.search('C18', k)],
    [k for k in all_gpeps.keys() if re.search('PGC', k)],
]

bp = pd.DataFrame()
pgc = pd.DataFrame()
c18 = pd.DataFrame()
files = get_files(r'.', exts=['.xlsx'])
for file in files:
    base = ntpath.basename(file)
    base, _ = ntpath.splitext(base)
    print(f'In file {base}:')
    bf = ByFile(file)
    bf.fill_no_glycans()
    bf.remove_reverse(modify=True)
    bf.determine_glycosites()
    bf.filter_hits(modify=True, score=100)
    bf.frame = bf.reduce_frame(gp_only=True)
    total = bf.total_gp()
    if re.search('BP', base):
        print('accessing BP')
        bp = combine_dfs(bp, total)
    elif re.search('C18', base):
        c18 = combine_dfs(c18, total)
    elif re.search('PGC', base):
        pgc = combine_dfs(pgc, total)
    t = create_ident(total, ['peptide', 'z'])

    print('\t\t\t', f'found {len(set(t))} glycopeptides in total')

```

```python
# number of glycans per backbone that are shared
c = {}
peps = []
g = {}
gpep = []
for d in [bp, c18, pgc]:
    data = d.clean_peptide.tolist()
    peps.extend(list(set(data)))
    gp = create_ident(d, ['clean_peptide', 'glycan'])
    gpep.extend(set(gp))
for p in peps:
    c[p] = c.get(p, 0) + 1
c = dict(sorted(c.items(), key=lambda x: x[1], reverse=True))

for gp in gpep:
    g[gp] = g.get(gp, 0) + 1
g = dict(sorted(g.items(), key=lambda x: x[1], reverse=True))

common = [k for k in c.keys() if c[k]==3]
g_common = [k for k in g.keys() if g[k]==3]

columns = ['C18', 'PGC', 'BP']
frames = [c18, pgc, bp]
glycs = []
source = []
peptides = []
for peptide in common:
    for i, frame in enumerate(frames):
        peptides.append(peptide)
        sub = frame[frame.clean_peptide==peptide]
        gly = sub.glycan.unique()
        glycs.append(len(gly))
        source.append(columns[i])


data = pd.DataFrame({'column':source, 'peptide':peptides, 'num gly':glycs})
domain = data.column.unique().tolist()
dom = {
    [d for d in domain if re.search('PGC', d)][0]: '#cc7833',
    [d for d in domain if re.search('BP', d)][0]: '#86858f',
    [d for d in domain if re.search('C18', d)][0]: '#43648c'
}
_range = ['#86858f', '#43648c', '#cc7833']


bars = alt.Chart(data).mark_boxplot().encode(
    x=alt.X('column', title='Column Type'),
    y=alt.Y('mean(num gly)', title='Glycan Count'),
    color=alt.Color('column:O', scale=alt.Scale(domain=list(dom.keys()), range=list(dom.values()))),
)

# error_bars = alt.Chart(data).mark_errorbar(extent='stdev').encode(
#     x=alt.X('column', title='Column Type'),
#     y=alt.Y('mean(num gly)', title='Glycan Count'),
# )
bars
```
